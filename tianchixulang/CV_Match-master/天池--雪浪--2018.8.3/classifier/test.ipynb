```python
import keras.backend as backend
import os
import cv2
import time
import numpy as np
import csv

start = time.time()

y_val_predict_all = []
y_val = []

csv_list=[]

#读取验证集normal的一张图像
for name in os.listdir("xuelang/test2"):
    img = cv2.imread("xuelang/test2/" + name)
    #转换为rgb
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#             batch_x = np.zeros((289,) + (224,224,3),backend.floatx())

    #batch_x = np.zeros((6,336,336,3),backend.floatx())
#     x = 0
#     for i in range(3):
#         y = 0
#         for j in range(2):
#             batch_x[i * 2 + j] = cv2.resize(img[y:y + 960, x:x + 960], (336, 336), interpolation=cv2.INTER_CUBIC)
#             y = y + 960
#         x = x + 800
    img = cv2.resize(img, (800, 600), interpolation=cv2.INTER_AREA)
    batch_x = np.zeros((1,600,800,3),backend.floatx())
    batch_x[0] = img


#             x = 0
#             for i in range(17):
#                 y = 0
#                 for j in range(17):
#                     batch_x[i * 17 + j] = img[y:y + 224, x:x + 224]
#                     y = y + 106
#                 x = x + 146
    #一个图片产生的数量 17*17
    #不打印    
    y_val_predict = model.predict(batch_x/255.,verbose=1,batch_size=8)
#             print(batch_x.shape)
#             print(batch_x)
#             print(y_val_predict.shape)
    #取其中可能性最大的为该图片的有问题与否可能性
    #print(y_val_predict)
    y_val_predict = np.max(y_val_predict,axis=0)[0]
    csv_list.append((name,y_val_predict))
    #print(y_val_predict)
    #normal 可能性应该为0
print("预测完成")





end = time.time()

print("用时:",end - start)
```
