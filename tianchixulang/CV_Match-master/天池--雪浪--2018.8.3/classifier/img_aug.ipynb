```python
import os
dict1={}
dict2={}
for img_name in os.listdir("../disk1/train/bad"):
    bad_type = img_name[:img_name.index('_')]
    if bad_type not in dict1:
        dict1[bad_type] = 0
    if bad_type not in dict2:
        dict2[bad_type] = []
    dict1[bad_type] += 1
    dict2[bad_type].append(img_name)
```

```python
#增强到5000
TRAIN_DIR = "../zp/xuelang/resize_data/bad/"
TRAIN_AUG_DIR = "../disk1/train_aug_600_800/bad/"
import cv2
import os
from random import choice
from imgaug import augmenters as iaa
import imgaug as ia

seq = iaa.Sequential([
            iaa.Fliplr(0.5),
            iaa.Flipud(0.5),
            #0.8,1.2
            iaa.Multiply((0.9, 1.1)),
            #iaa.AddToHueAndSaturation((-30, 45)),
        ])

#稀少类别的增强到200张
for v,k in dict1.items():
    if k < 200:
        j = 0
        imageList = []
        #做200-k次增强
        while j < 200 - k:
            image = cv2.imread(TRAIN_DIR + choice(dict2[v]))
            imageList.append(image)
            
            j += 1
            
        images_aug = seq.augment_images(imageList)
        for n in range(len(images_aug)):
            cv2.imwrite(TRAIN_AUG_DIR + v + "_aug_" + str(n) + ".jpg",images_aug[n])
    print(v+"完成")
```
