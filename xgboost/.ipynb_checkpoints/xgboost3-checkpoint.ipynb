```python
import warnings
warnings.filterwarnings("ignore")
import  numpy as np
import pandas as pd
from pandas import datetime
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# statistics
from statsmodels.distributions.empirical_distribution import ECDF

# time series analysis
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from fbprophet import Prophet
```

```python
train = pd.read_csv("train.csv" ,parse_dates = True , low_memory = False ,index_col = 'Date')
store = pd.read_csv("store.csv",low_memory = False)
train.index
```

```python
print("In total:" ,train.shape)
train.head(5)
```

```python
train['Year'] = train.index.year
train['Month'] = train.index.month
train['Day'] = train.index.day
train['WeekOfYear'] = train.index.weekofyear
train['SalePerCustomer'] = train['Sales'] / train['Customers']
train['SalePerCustomer'].describe()
```

```python
sns.set(style = "ticks")
c = "#386B7F"
plt.figure(figsize = (12 ,6))

plt.subplot(311)
cdf = ECDF(train['Sales'])   #ECDF经验分布函数值
plt.plot(cdf.x , cdf.y ,label = "statmodels" , color = c)
plt.xlabel('Sales')
plt.ylabel('ECDF')

plt.subplot(312)
cdf = ECDF(train['Customers'])   #ECDF经验分布函数值
plt.plot(cdf.x , cdf.y ,label = "statmodels" , color = c)
plt.xlabel('Customers')

plt.subplot(313)
cdf = ECDF(train['SalePerCustomer'])   #ECDF经验分布函数值
plt.plot(cdf.x , cdf.y ,label = "statmodels" , color = c)
plt.xlabel('SalePerCustomer')
```

很明显，这些信息在数据中是缺失的。没有特定的模式。在这种情况下，用中间值替换NaN是完全有意义的（这比平均值少了两倍）

继续使用丢失的数据。Promo2SinceWeek呢?我们是否可以观察到不一致的数据点？
对于是否有Promo数据处理

```python
promo = store[pd.isnull(store.Promo2SinceWeek)]
promo[promo.Promo2 != 0].shape
```

对于推广活动没有数据用0填充，对于是否有competition （CompetitionOpenSinceMonth and CompetitionOpenSinceYear）也可以同样的处理

```python
#将缺失值用0代替
store.fillna(0 , inplace = True)
```

```python
print("joining train set with an additional store information")
train_store = pd.merge(train , store ,how = 'inner' ,on = 'Store') #以Store为key合拼两张表
#how='inner' 参数指的是当左右两个对象中存在不重合的键时，
#取结果的方式：inner 代表交集；outer 代表并集；left 和 right 分别为取一边。
print("In total:" ,train_store.shape) 
train_store.head()
```

完成我们的初步数据分析,我们可以添加一些变量来描述竞争和促销活动的时间

```python
#完全开放的时间（在各个月份）
train_store['CompetitionOpen'] = 12 * (train_store.Year - train_store.CompetitionOpenSinceYear)+\
    (train_store.Month - train_store.CompetitionOpenSinceMonth)
#优惠开放时间
train_store['PromoOpen'] = 12 * (train_store.Year - train_store.Promo2SinceYear) + \
    (train_store.WeekOfYear - train_store.Promo2SinceWeek) / 4.0
#用0代替空值
train_store.fillna(0 , inplace =True)
#平均每个商店类型的促销时间和竞争开放时间
train_store.loc[: ,['StoreType', 'Sales', 'Customers', 'PromoOpen', 'CompetitionOpen']].groupby('StoreType').mean()
```

最畅销和最拥挤的是StoreType A 似乎并不是最容易受到竞争对手影响的，相反,它是一个StoreType B，这也是最长的促销期

相关性分析
我们已经完成了向数据添加新变量的工作，所以现在我们可以通过绘制seaborn heatmap来检查总体相关性

```python
#排除开放变量
corr_all = train_store.drop('Open', axis = 1).corr()
#为上三角生成一个掩码
mask = np.zeros_like(corr_all, dtype = np.bool)
mask[np.triu_indices_from(mask)] = True
#设置matplotlib图形
f, ax = plt.subplots(figsize = (11, 9))
#用蒙版和正确的纵横比来绘制热图
sns.heatmap(corr_all, mask = mask,
            square = True, linewidths = .5, ax = ax, cmap = "BuPu")      
plt.show()
```

正如前面提到的，我们在商店的销售数量和顾客之间有着很强的正相关关系。我们还可以观察到商店有一个正在进行的促销活动（Promo等于1）和顾客数量之间的正相关。
然而，一旦商店继续连续促销(Promo2 equal to 1) Customers和Sales似乎保持不变甚至减少，这是在热图上的负相关。在商店的促销活动和一周的一天之间也存在着同样的负相关

时间序列分析对于每一个商店

为什么时间序列与回归问题的不同？

他是与时间相关的，而线性回归问题假设各个特征是独立的

随着趋势的增加或是减少，更多的时间序列有某种形式的季节性趋势，像如:特定时间框架的变化，像如在这些数据中圣诞节

我们建立了一个关于商店类型的时间序列分析而不是单独的商店，这种方法的主要优点是它的简单表示和数据集的不同趋势和季节性的总体描述。

在这一节中，我们将分析时间序列数据：它的趋势、单值和自相关。通常在分析的最后，我们能够开发一个季节性的ARIMA（自回归集成移动平均）模型，但它不会是我们今天的主要关注点。相反，我们试着去理解数据，然后用先知方法来做预测



季节性分析

我们从商店类型中选取四家商店来代表他们的分组：

Store number 2 for StoreType A

Store number 85 for StoreType B,

Store number 1 for StoreType C

Store number 13 for StoreType D.

使用resample方法对数据进行采样，以更清楚地看到当前趋势，这也是有意义的。



```python
#输入float类型的数据
train['Sales'] = train['Sales'] * 1.0
#店铺类型
sale_a = train[train.Store == 2]['Sales']
sale_b = train[train.Store == 85]['Sales'].sort_index(ascending = True)
sale_c = train[train.Store == 1]['Sales']
sale_d = train[train.Store == 13]['Sales']

f ,(ax1 ,ax2 ,ax3,ax4) = plt.subplots(4 , figsize = (12 ,13))

sale_a.resample('W').sum().plot(color = c,ax = ax1)
sale_b.resample('W').sum().plot(color = c,ax = ax2)
sale_c.resample('W').sum().plot(color = c,ax = ax3)
sale_d.resample('W').sum().plot(color = c,ax = ax4)
```

零售销售对于StoreType A和C 在圣诞节的时候会达到顶峰然后在假期结束后下降,我们可能已经看到了同样的趋势对于StoreType D（在底部）但是从7月起就没有信息了从2014年7月到2015年1月，这些商店被关闭

# Yearly trend

接下来要检查的是一系列的趋势

```python
f ,(ax1 ,ax2 , ax3 ,ax4) = plt.subplots(4 , figsize = (12 ,13))
#monthly
decomposotion_a = seasonal_decompose(sale_a , model = 'additive' , freq = 365)
decomposotion_a.trend.plot(color = c ,ax = ax1)

decomposotion_b = seasonal_decompose(sale_b , model = 'additive' , freq = 365)
decomposotion_b.trend.plot(color = c ,ax = ax2)

decomposotion_c = seasonal_decompose(sale_c , model = 'additive' , freq = 365)
decomposotion_c.trend.plot(color = c ,ax = ax3)

decomposotion_d = seasonal_decompose(sale_d , model = 'additive' , freq = 365)
decomposotion_d.trend.plot(color = c ,ax = ax4)
```

整体销售额似乎在增长,但是不是StoreType C  （三分之一）,即使在StoreType A 是数据集中最畅销的商店类型似乎与StoreType C 有着相同的规律

# Autocorrelaion

我们的时间序列分析的下一步是回顾自相关函数（ACF）和部分自相关函数（PACF）图：

ACF是一种度量timeseries与它自身的滞后版本之间的相关性的度量。例如，在延迟5时，ACF将在瞬间“t1”。“tn”与系列的“t1-5”。“tn-5”（t1-5和tn为终点）进行比较。

另一方面，PACF则度量timeseries与它的滞后版本之间的相关性，但是在消除了中间比较所解释的变化之后。如。在延迟5中，它会检查相关性，但是去掉了滞后1到4的效果

```python
plt.figure(figsize = (12 ,8))
#a
plt.subplot(421);plot_acf(sale_a , lags = 50 ,ax = plt.gca() , color =c)
plt.subplot(422);plot_pacf(sale_a , lags = 50 ,ax = plt.gca() , color =c)
#b
plt.subplot(423);plot_acf(sale_b , lags = 50 ,ax = plt.gca() , color =c)
plt.subplot(424);plot_pacf(sale_b , lags = 50 ,ax = plt.gca() , color =c)
#c
plt.subplot(425);plot_acf(sale_c , lags = 50 ,ax = plt.gca() , color =c)
plt.subplot(426);plot_pacf(sale_c , lags = 50 ,ax = plt.gca() , color =c)
#d
plt.subplot(427);plot_acf(sale_d , lags = 50 ,ax = plt.gca() , color =c)
plt.subplot(428);plot_pacf(sale_d , lags = 50 ,ax = plt.gca() , color =c)

plt.show()
```

我们可以水平地阅读这些图,每一个水平对是一个 StoreType(从A到D),普遍的，一般来说，这些图显示的是级数的相关性，滞后的时间单元x级数与自身的相关性，在x时间单位上落后。

每一对情节都有两个共同点：非随机的时间序列和高 lag-1 这可能需要更高阶的差异 d/D）
